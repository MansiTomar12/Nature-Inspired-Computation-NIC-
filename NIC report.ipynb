```python
#BPP1
# In problem BPP1, b = 10 and the weight of item ùëñ (where ùëñ starts at 1 and finishes at 500) will be ùíä/ùüê.
import numpy as np
import random
#Calculating the fitness of the solution
def calc_fitness(sol):
    d = {}
    for i in range(1,11):
        d[i] = 0
    for index in range(len(sol)):
        d[sol[index]] += (index+1)/2
    return max(d.values()) - min(d.values())

#binary tournament selection - Randomly choose a chromosome from the population; 
def binary_tournament_selection(Population, a,b):
    idx = np.random.randint(a, size=b)
    return Population[idx,:][0]

#Single point crossover -Randomly select a ‚Äòcrossover point‚Äô which should be smaller than the total length of the chromosome.
def single_point_crossover(Population):
    a = binary_tournament_selection(Population,10,1)
    b = binary_tournament_selection(Population,10,1) 
    x=np.random.randint(1,500)
    child1=list(a[:x])+list(b[x:])
    child2=list(b[:x])+list(a[x:])
    
    return (child1,child2)

#Multi gene mutation-: You should use the following parameterised mutation operator for this, Mk, where k is an integer
def multi_gene_mutation(child1,child2,k):
    for i in range(k):
        ran=random.randint(1,len(child1))
        val=random.randint(1,10)
        child1[ran-1]=val
        #ran=random.randint(1,len(child1))
        #val=random.randint(1,10)
        child2[ran-1]=val   
    return (child1,child2)
    
#weakest solution replacement -: If the new solution is fitter than the worst in the population, then overwrite theworst with the new solution.
def weakest_solution(child, population, fitness_list):
    c1=calc_fitness(child)
    max_list=[]
    Max_val=max(fitness_list)
    if c1<=Max_val:
        for i in range(len(fitness_list)):
            if fitness_list[i]==Max_val:
                max_list.append(i)
        index = np.random.randint(0, len(max_list), size=1)[0]
        fitness_list[max_list[index]] = c1
        population[max_list[index]] = child

            
    return population, fitness_list
```
